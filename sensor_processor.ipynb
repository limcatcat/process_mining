```python
import pandas as pd
```

```python
df_dist = pd.DataFrame()

for i in range(1, 31):
    if i < 10:
        i = '0' + str(i)
    file_path = f'../scene{i}/scene{i}_distance_sensor_laptop_shelf.csv'

    with open(file_path, 'r') as f:
        df = pd.read_csv(file_path)

    df_dist = pd.concat([df_dist, df], ignore_index=True)
```

```python
df_dist
```

```python
df_dist['diff'] = df_dist['distance_in_cm'].diff()
```

```python
df_dist
```

```python
print(df_dist['diff'].min())
print(df_dist['diff'].max())

print(df_dist['diff'].abs().min())
print(df_dist['diff'].abs().max())
```

```python
df_dist['diff_abs'] = df_dist['diff'].abs()
```
